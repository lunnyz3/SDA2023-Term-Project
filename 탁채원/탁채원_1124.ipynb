### 결측치 처리 방법

```python
LG_pre['OPS'].isnull().sum()
```

```python
LG_pre_OPS_missing = LG_pre[LG_pre['OPS'].isnull()]
LG_pre_OPS_missing
```

```python
# 1. 선형보간법 이용
df = LG_pre[['OPS', 'year', 'batter_name']]
df = df.interpolate()
df_OPS = df.groupby('year')['OPS'].mean().reset_index()
df_OPS = df_OPS.sort_values(by='year')
df_OPS
```

```python
# 2. 타자의 평균 OPS로 대체
df2 = pre_season.copy()

df2['OPS'].fillna(df2.groupby('batter_name')['OPS'].transform('mean'), inplace=True)

배정대 = df2[df2['batter_name'] == '배정대']
배정대
```

```python
df2_LG = df2.groupby('year')['OPS'].mean().reset_index()
df2_LG.rename(columns = {'OPS':'regular'},inplace=True)
df2_LG = df2_LG.sort_values(by='year')
df2_LG
```

```python
# 3. 타자의 다른 특성으로 예측
df3 = LG_pre.copy()

from sklearn.linear_model import LinearRegression
import numpy as np


for id in df3['batter_id'].unique():
    player_data = df3[df3['batter_id'] == id]
    train_data = player_data.dropna(subset=['OPS', 'SLG', 'OBP'])
    
    if not train_data.empty:
        X = train_data[['SLG', 'OBP']]
        y = train_data['OPS']
        
        model = LinearRegression()
        model.fit(X, y)
        
        test_data = player_data[pd.isnull(player_data['OPS'])][['SLG', 'OBP']]
        if not test_data.empty:
            predicted_ops = model.predict(test_data)
            df3.loc[test_data.index, 'OPS'] = predicted_ops

df3['OPS'].isnull().sum()
```

```python
df3 = df3.groupby('year')['OPS'].mean().reset_index()
df3.rename(columns = {'OPS':'regular'},inplace=True)
df3 = df3.sort_values(by='year')
df3
```

```python
# 4. 타자의 과거 성적으로 예측
df4 = pre_season.copy()
df4.sort_values(['batter_id', 'year'], inplace=True)

for player_id in df4['batter_id'].unique():
    player_data = df4[df4['batter_id'] == player_id]
    years_available = player_data['year'].nunique()
    train_data = player_data.dropna(subset=['OPS'])
    
    if not train_data.empty:
        n = min(3, years_available)  
        train_data['과거평균OPS'] = train_data['OPS'].shift(1).rolling(window=n).mean()
        
        test_data = player_data[pd.isnull(player_data['OPS'])]
        if not test_data.empty:
            df4.loc[test_data.index, 'OPS'] = test_data['과거평균OPS']

df4_LG = df4.groupby('year')['OPS'].mean().reset_index()
df4_LG.rename(columns = {'OPS':'regular'},inplace=True)
df4_LG = df4_LG.sort_values(by='year')
df4_LG
```
